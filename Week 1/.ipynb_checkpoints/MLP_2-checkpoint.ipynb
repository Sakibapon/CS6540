## Learning Objective

Explain why MLPs fail for image data by comparing parameter counts and demonstrating the scalability problem with concrete calculations.

## 1. Historical Context

> Before deep learning, computer vision relied on hand-crafted features like SIFT, HOG, and SURF. The natural question arose: could neural networks learn features directly from raw pixels? Early attempts using Multi-Layer Perceptrons revealed a fundamental scalability problem that would ultimately motivate the development of Convolutional Neural Networks.

### The Core Problem

MLPs treat images as flat vectors, requiring:
- **Every pixel** to connect to **every neuron** in the first hidden layer
- This creates $O(\text{pixels} \times \text{neurons})$ parameters
- Parameter count explodes with image resolution

## 2. MLP Architecture for Image Classification

Consider classifying images using a standard MLP with two hidden layers:

**Architecture:**
```
Input (W × H × C pixels, flattened)
  ↓
FC1: 1,000 neurons
  ↓ ReLU
FC2: 1,000 neurons  
  ↓ ReLU
Output: N classes
```

**Parameter Formula (weights only):**

$$\text{Parameters} = (W \times H \times C) \times 1000 + 1000 \times 1000 + 1000 \times N$$

````python
def count_mlp_parameters(input_size, hidden1, hidden2, output_size):
    """
    Count parameters in a 2-hidden-layer MLP (weights only).
    
    Parameters: input×h1 + h1×h2 + h2×output
    """
    l1 = input_size * hidden1   # Input → Hidden1
    l2 = hidden1 * hidden2      # Hidden1 → Hidden2
    l3 = hidden2 * output_size  # Hidden2 → Output
    return l1 + l2 + l3, l1, l2, l3
````

## 3. Case Study: Fashion MNIST

**Dataset:** 28×28 grayscale images, 10 clothing categories

**Architecture:**
```
Input: 28 × 28 × 1 = 784 pixels
  ↓
FC1: 784 × 1,000 = 784,000 params
  ↓
FC2: 1,000 × 1,000 = 1,000,000 params
  ↓
Output: 1,000 × 10 = 10,000 params
```

````python
# Fashion MNIST: 28×28 grayscale
INPUT_FASHION = 28 * 28 * 1  # 784 pixels

total, l1, l2, l3 = count_mlp_parameters(INPUT_FASHION, 1000, 1000, 10)

print("Fashion MNIST Classification (MLP)")
print("=" * 45)
print(f"Input:  28 × 28 × 1 = {INPUT_FASHION:,} pixels")
print(f"\nParameter Breakdown:")
print(f"  FC1: {INPUT_FASHION:,} × 1,000 = {l1:>12,}")
print(f"  FC2: 1,000 × 1,000   = {l2:>12,}")
print(f"  Out: 1,000 × 10      = {l3:>12,}")
print(f"  ─────────────────────────────────")
print(f"  Total:               = {total:>12,}")
print(f"\n≈ {total/1e6:.2f} Million parameters")
````

## 4. The Scalability Problem

What happens when we increase image resolution?

### 64×64 RGB Image

**Architecture:**
```
Input: 64 × 64 × 3 = 12,288 pixels
  ↓
FC1: 12,288 × 1,000 = 12,288,000 params
  ↓
FC2: 1,000 × 1,000 = 1,000,000 params
  ↓
Output: 1,000 × 2 = 2,000 params
```

````python
# 64×64 RGB image (from lecture slides)
INPUT_64 = 64 * 64 * 3  # 12,288 pixels

total_64, l1_64, l2_64, l3_64 = count_mlp_parameters(INPUT_64, 1000, 1000, 2)

print("64×64 RGB Classification (MLP)")
print("=" * 45)
print(f"Input:  64 × 64 × 3 = {INPUT_64:,} pixels")
print(f"\nParameter Breakdown:")
print(f"  FC1: {INPUT_64:,} × 1,000 = {l1_64:>12,}")
print(f"  FC2: 1,000 × 1,000    = {l2_64:>12,}")
print(f"  Out: 1,000 × 2        = {l3_64:>12,}")
print(f"  ─────────────────────────────────")
print(f"  Total:                = {total_64:>12,}")
print(f"\n≈ {total_64/1e6:.2f} Million parameters")
````

### 2000×2000 RGB Image (High Resolution)

**Architecture:**
```
Input: 2000 × 2000 × 3 = 12,000,000 pixels
  ↓
FC1: 12,000,000 × 1,000 = 12,000,000,000 params
  ↓
FC2: 1,000 × 1,000 = 1,000,000 params
  ↓
Output: 1,000 × 2 = 2,000 params
```

````python
# 2000×2000 RGB image (from lecture slides)
INPUT_2000 = 2000 * 2000 * 3  # 12,000,000 pixels

total_2000, l1_2000, l2_2000, l3_2000 = count_mlp_parameters(INPUT_2000, 1000, 1000, 2)

print("2000×2000 RGB Classification (MLP)")
print("=" * 50)
print(f"Input:  2000 × 2000 × 3 = {INPUT_2000:,} pixels")
print(f"\nParameter Breakdown:")
print(f"  FC1: {INPUT_2000:,} × 1,000 = {l1_2000:>15,}")
print(f"  FC2: 1,000 × 1,000        = {l2_2000:>15,}")
print(f"  Out: 1,000 × 2            = {l3_2000:>15,}")
print(f"  ───────────────────────────────────────")
print(f"  Total:                    = {total_2000:>15,}")
print(f"\n≈ {total_2000/1e9:.2f} BILLION parameters")
````

## 5. Comparison Summary

````python
import matplotlib.pyplot as plt

# Compare all three cases
cases = [
    ("Fashion MNIST\n28×28×1", INPUT_FASHION, total),
    ("Small RGB\n64×64×3", INPUT_64, total_64),
    ("HD Photo\n2000×2000×3", INPUT_2000, total_2000),
]

print("\nMLP Parameter Scaling Summary")
print("=" * 55)
print(f"{'Image':<20} {'Input Size':>15} {'Parameters':>18}")
print("-" * 55)
for name, inp, params in cases:
    name_clean = name.replace('\n', ' ')
    if params >= 1e9:
        p_str = f"{params/1e9:.2f} Billion"
    else:
        p_str = f"{params/1e6:.2f} Million"
    print(f"{name_clean:<20} {inp:>15,} {p_str:>18}")
print("=" * 55)

# Visualization
fig, ax = plt.subplots(figsize=(10, 5))
names = [c[0] for c in cases]
params = [c[2] for c in cases]
colors = ['#2ecc71', '#f39c12', '#e74c3c']

bars = ax.bar(names, [p/1e9 for p in params], color=colors, edgecolor='white', linewidth=2)
ax.set_ylabel('Parameters (Billions)', fontsize=12, fontweight='bold')
ax.set_title('MLP Parameter Count vs Image Size', fontsize=14, fontweight='bold')
ax.axhline(y=1, color='red', linestyle='--', alpha=0.7, label='1 Billion')

# Add value labels
for bar, p in zip(bars, params):
    label = f"{p/1e9:.2f}B" if p >= 1e9 else f"{p/1e6:.1f}M"
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3, 
            label, ha='center', fontweight='bold', fontsize=11)

ax.legend()
plt.tight_layout()
plt.show()
````

## 6. Why This Matters

| Problem | 64×64 Image | 2000×2000 Image |
|---------|-------------|------------------|
| **Parameters** | ~13M | ~12B |
| **Memory (Training)** | ~0.15 GB | ~135 GB |
| **Practical?** | Yes | No |

### Key Issues:

1. **Memory**: 12 billion parameters × 4 bytes × 3 (gradients + optimizer) ≈ 135 GB
2. **Overfitting**: With 10K training images, that's 1.2M parameters per example
3. **No Spatial Awareness**: Adjacent pixels are treated as independent features

## 7. The Solution: Convolutional Neural Networks

> The scalability problem of MLPs directly motivated the development of CNNs. By using local connectivity and weight sharing, CNNs reduce parameters by orders of magnitude while preserving spatial structure.

**CNN Approach:**
```
Conv Layer: 3×3 kernel, 32 filters
  Parameters: 3 × 3 × 3 × 32 = 864 (not 12M!)
```

**Key Difference:**
- MLP: Every pixel → every neuron
- CNN: Small region → each filter (weights shared across image)

````python
# Quick comparison: MLP vs CNN for same image
mlp_first_layer = INPUT_2000 * 1000  # 12 billion
cnn_first_layer = 3 * 3 * 3 * 32     # 864

print("First Layer Parameter Comparison (2000×2000 RGB)")
print("=" * 50)
print(f"MLP FC1:  {INPUT_2000:,} × 1,000 = {mlp_first_layer:,}")
print(f"CNN Conv1: 3 × 3 × 3 × 32      = {cnn_first_layer:,}")
print(f"\nReduction: {mlp_first_layer/cnn_first_layer:,.0f}× fewer parameters")
````

## Summary

**Key Numbers (from lecture):**
- 64×64×3 image with MLP → ~13.38 Million parameters
- 2000×2000×3 image with MLP → ~12.01 Billion parameters

**Formula:**
$$\text{MLP Parameters} = (W \times H \times C) \times H_1 + H_1 \times H_2 + ...$$

**Conclusion:** The quadratic scaling of MLP parameters with image dimensions makes them impractical for real-world image classification. This limitation led to the development of CNNs, which we cover next.
